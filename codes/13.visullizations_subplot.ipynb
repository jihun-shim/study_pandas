## visuallization

### matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x_data_first = [1,2,3,4]
y_data_first = [1,2,3,4]
x_data_second = [2,3,1,1]
y_data_second = [2,3,1,1]
x_data_third = [1,2,3,4]
y_data_third = [2,1,4,1]
x_data_fourth = [4,3,2,1]
y_data_fourth = [4,3,2,1]
```

### 정렬
- add_subplot, subplots()

```python
#기본구성

# fig, axes = plt.subplots(2,2)
fig, axes = plt.subplots(1,2)
# fig, axes = plt.subplots(2,2, figsize=(14,6))

axes[0].plot(x_data_first, y_data_first)
axes[0].plot(x_data_third, y_data_third)

plt.show()
```

### 구성요소 추가

```python

# fig, axes = plt.subplots(2,2)
fig, axes = plt.subplots(1,2)
# fig, axes = plt.subplots(2,2, figsize=(14,6))

# fig 에 타이틀 달기
fig.suptitle('Super Title')

axes[0].plot(x_data_first, y_data_first, label='legend first') #레이블 포함
axes[0].plot(x_data_third, y_data_third, label='legend second')
axes[0].set_title('axes first') # subplots 는 set이 붙는다
axes[0].set_xlabel('first x_data')
axes[0].legend() #legend는 그냥쓰기

axes[1].plot(x_data_third, y_data_third)
axes[1].set_title('axes second')
axes[1].set_ylabel('second y_data')

plt.tight_layout() #그래프에 겹치는 글자 자동으로 떨어트리기
plt.show()
```

### figure size 조정

```python
fig, axes = plt.subplots(1,2, figsize=(10,6), dpi = 100) #그래프 칸 갯수 및 크기 지정 
# fig, axes = plt.subplots(2,2, figsize=(14,6))

# fig 에 타이틀 달기
fig.suptitle('Super Title')

axes[0].plot(x_data_first, y_data_first, label='legend first') #레이블 포함
axes[0].plot(x_data_third, y_data_third, label='legend second')
axes[0].set_title('axes first') # subplots 는 set이 붙는다
axes[0].set_xlabel('first x_data')
axes[0].legend() #legend는 그냥쓰기

axes[1].plot(x_data_third, y_data_third)
axes[1].set_title('axes second')
axes[1].set_ylabel('second y_data')

plt.tight_layout() #그래프에 겹치는 글자 자동으로 떨어트리기
plt.show()
```
