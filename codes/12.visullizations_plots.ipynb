```python
import matplotlib.pyplot as plt
```

#### 기본 구성

```python
x_data_first = [1,2,3,4]
y_data_first = [1,2,3,4]
x_data_second = [2,3,1,1]
y_data_second = [2,3,1,1]
x_data_third = [1,2,3,4]
y_data_third = [2,1,4,1]
x_data_fourth = [4,3,2,1]
y_data_fourth = [4,3,2,1]
```

### 단일변수 표현

```python
plt.plot(x_data_first) # 리스트로 저장되어 있기 때문에 단일변수 표현 즉 인덱스가 있기 때문
plt.show() 
```

### 다변수 표현

```python
plt.plot(x_data_first, y_data_first)
plt.plot(y_data_third, x_data_third)

plt.show()
```

### 설명 구성

```python
plt.plot(x_data_first, y_data_first, label='x_data_legend') # 범례
plt.plot(y_data_third, x_data_third, label='y_data_legend')

plt.title('Visuallization With Title') #타이틀
plt.xlabel('x_data') # x축 레이블
plt.ylabel('y_data') # y축 레이블
plt.legend() 


plt.show()
```

#### 한글 처리

```python
import matplotlib as mpl 
# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
plt.plot(x_data_first, y_data_first, label='x_data_범례') #범례
plt.plot(y_data_third, x_data_third, label='y_data_범례')

plt.title('Visuallization With 타이틀') #타이틀
plt.xlabel('x_데이터') # x축 레이블
plt.ylabel('y_데이터') # y축 레이블
plt.legend() #범례

plt.savefig('visualliztions_plots.png') # 장소 지정 및 저장할 형태
plt.show()
```

### figure 넓이, 높이 조정

```python
# plt.figure(figsize=(10,6)) # 10 * 6인치 # figure 넓이, 높이를 조정 단위는 인치
plt.figure(figsize=(10,6), dpi=100) # dpi 는 픽셀로 표현  

plt.plot(x_data_first, y_data_first, label='x_data_범례') #범례
plt.plot(y_data_third, x_data_third, label='y_data_범례')

plt.title('Visuallization With 타이틀') #타이틀
plt.xlabel('x_데이터') # x축 레이블
plt.ylabel('y_데이터') # y축 레이블
plt.legend() #범례

plt.savefig('visualliztions_plots.png') # 장소 지정 및 저장할 형태
plt.show()
```
