```python
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df_first = pd.DataFrame(data=data, columns=columns)
df_first
```

### 월 단위 resampling

```python
# df_resample_first.resample('MS').first() # MS = 그달의 첫 날, ME = 그달의 마지막날 로 묶는다 어떠한 것을 대표 값으로 쓸 것인지
df_resample_first.resample('ME').first()
```

```python
# how = {
#     '시가' : 'first'
#     , '종가' : 'last'
#     , '고가' : max
#     , '저가' : min
#     , '거래량' : sum
# }

import numpy as np
how = {
    '시가' : 'first'
    , '종가' : 'last'
    , '고가' : np.max
    , '저가' : np.min
    , '거래량' : np.sum
    , '등락률' : np.mean
}
df_resample_first.resample('ME').apply(how)
```

#### 일 단위

```python
df_resample_third = df_resample_first.resample('3D').apply(how)
df_resample_third.head()

# resample('3D')**는 3일 단위로 데이터를 리샘플링합니다.

# apply(how):
# apply() 함수는 각 그룹에 대해 사용자가 정의한 함수(여기서는 how)를 적용하는 함수입니다.
# **how**는 이 코드에서 사용자가 지정한 함수를 의미합니다. 
# 이 함수는 3일 간격으로 그룹화된 데이터에 대해 어떻게 처리할지를 정의합니다.
# 예를 들어, 평균값을 계산하거나, 최대값을 구하는 등의 작업을 할 수 있습니다.
```

```python
df_resample_first.resample('12H').apply(how) #시간으로 나타냄
```

### 수익율

```python
data = {'삼성전자': [52200, 52300, 52900, 52000, 51700], 
           'LG전자': [68200, 67800, 68800, 67500, 66300]}
```

```python
df_first = pd.DataFrame(data)
```

```python
# pd.DataFrame(data).pct_change(periods = 2)
df_first.pct_change(periods = 2)

# pct_change():
# pct_change()는 비율 변화(percentage change)를 계산하는 함수입니다. 즉, 현재 값과 이전 값의 차이를 퍼센트로 계산합니다.

# periods=2:
# periods=2는 2개의 이전 값을 기준으로 변화율을 계산하라는 의미입니다.
# 즉, 각 값은 그 자신을 제외한 2개 이전 값과 비교하여 변화율을 계산하게 됩니다.
```

```python
df_first.cumprod() #누적곱 값
```
