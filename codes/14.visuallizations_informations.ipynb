```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정 (NanumGothic) 
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
# API 사용
import FinanceDataReader as fdr
symbol_kospi = fdr.DataReader('KS11','2023') #가져올 데이터와 년도 설정
symbol_kospi
```

### 꺾은 선 그래프 표시

```python
plt.figure(figsize=(10,6), dpi = 100)

plt.plot(symbol_kospi['Close'], label='Close')
plt.plot(symbol_kospi['Open'], label='Open')

plt.title('2023년 이후 코스피 지수')
plt.xlabel('날짜')
plt.ylabel('지수')
plt.legend()

plt.show()
```
