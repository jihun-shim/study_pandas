### 꺾은 선 그래프 표시

```python
plt.figure(figsize=(10,6), dpi = 100)

plt.plot(symbol_kospi['Close'], label='Close')
plt.plot(symbol_kospi['Open'], label='Open')

plt.title('2023년 이후 코스피 지수')
plt.xlabel('날짜')
plt.ylabel('지수')
plt.legend()

plt.show()
```

```python
symbol_tway = fdr.DataReader(symbol='091810',start='2021', end='2022')
symbol_samsung = fdr.DataReader(symbol='005930',start='2021', end='2022')
```

```python
symbol_tway.shape,symbol_samsung.shape
```

```python
plt.plot(symbol_samsung.index, symbol_samsung['Close'])
plt.plot(symbol_tway.index, symbol_tway['Close'])

plt.show()
```

```python
# type(symbol_samsung['Close']) 타입확인
# symbol_samsung['Close'] 시리즈 확인
# symbol_samsung['Close'].iloc[0] 특정위치 값 확인

#지수화 *극과극인 값 비교가 어려운 대상과의 추세를 비교해보고 싶을때
serise_samsung = (symbol_samsung['Close'] / symbol_samsung['Close'].iloc[0]) * 100
serise_samsung
```

```python
serise_tway = (symbol_tway['Close'] / symbol_tway['Close'].iloc[0]) * 100
serise_tway
```

```python
plt.plot(symbol_samsung.index, serise_samsung)
plt.plot(symbol_tway.index, serise_tway)
plt.title('지수화 비교')
plt.show()
```
