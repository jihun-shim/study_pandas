## load from DB

```python
find_data = phone_collection.find()
type(find_data)
```

```python
phone_data = list(find_data)
phone_data
```

```python
import pandas as pd
```

```python
df_phone_list = pd.DataFrame(phone_data)
```

```python
df_phone_list.info()
```

```python
df_phone_list.describe(include='all')
```

## save to DB

```python
# type(df_phone_list)
df_phone_list_without_id = df_phone_list.drop('_id', axis=1)
df_phone_list_without_id
```

```python
data_to_insert = df_phone_list_without_id.to_dict(orient='records')
data_to_insert
```

```python
result = phone_collection.insert_many(data_to_insert)
```
