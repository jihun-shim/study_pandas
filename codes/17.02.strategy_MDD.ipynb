## 최대낙폭 (MDD) 
- Maximun Drawdown
- MDD = (1 - (종가 / 전고점)) * 100  #전고점은 이전 날 n전날의 고점

```python

# API 사용
import FinanceDataReader as fdr
symbol_kospi = fdr.DataReader('KS11','2000','20231231') #가져올 데이터와 년도 설정
symbol_kospi.tail() # tail은 head의 반대
```

```python
symbol_kospi['전고점'] = symbol_kospi['Close'].cummax()
symbol_kospi

# cummax()는 누적 최대값(cumulative maximum)을 계산하는 함수입니다.
```

```python
symbol_kospi['MDD'] = (1 - (symbol_kospi['Close'] / symbol_kospi['전고점'])) * 100 
symbol_kospi
```

```python
symbol_kospi['MDD'].max()
```

### Chart 작성

```python
import matplotlib.pyplot as plt
plt.rc('axes', unicode_minus = False)
```

```python
fig, axes = plt.subplots(2,1, figsize=(14,6))

# 종가 기준 그래프
axes[0].plot(symbol_kospi.index, symbol_kospi['Close'])

# MDD 
axes[1].plot(symbol_kospi.index, symbol_kospi['MDD']*-1)
axes[1].fill_between(symbol_kospi.index, symbol_kospi['MDD']*-1, alpha = 0.1, color = 'red')
# axes[1].plot(symbol_kospi.index, symbol_kospi['MDD'])
# axes[1].fill_between(symbol_kospi.index, symbol_kospi['MDD'], alpha = 0.1, color = 'red')

plt.tight_layout()
plt.show()
```
