## 평균, 중앙값과 첨도와 관계

```python
# 데이터 생성 
data_symmetric = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 대칭 분포 
data_positive_skew = [1, 2, 2, 3, 3, 3, 10, 10, 50] # 오른쪽 꼬리 
data_negative_skew = [-50, -10, -10, -3, -3, -3, -2, -2, -1] # 왼쪽 꼬리
```

```python
import numpy as np
from scipy.stats import skew, kurtosis # 왜도 계산 함수
```

```python
mean = np.mean(data_positive_skew)
median = np.median(data_positive_skew)
skew = skew(data_positive_skew)     # skew > 0 -> 오른쪽 꼬리
mean, median, skew
```
