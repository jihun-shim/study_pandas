```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'
```

```python
import FinanceDataReader as fdr
symbol_kospi = fdr.DataReader(symbol = 'KS11', start='1990', end='2023')
symbol_IXIC = fdr.DataReader(symbol = 'IXIC', start='1990', end='2023')
symbol_SSEC = fdr.DataReader(symbol = 'SSEC', start='1990', end='2023')
symbol_HSI = fdr.DataReader(symbol = 'HSI', start='1990', end='2023')
symbol_N225 = fdr.DataReader(symbol = 'N225', start='1990', end='2023')
symbol_FTSE = fdr.DataReader(symbol = 'FTSE', start='1990', end='2023')
symbol_FCHI = fdr.DataReader(symbol = 'FCHI', start='1990', end='2023')
symbol_GDAXI = fdr.DataReader(symbol = 'GDAXI', start='1990', end='2023')
```

```python
series_kospi = (symbol_kospi['Close']/symbol_kospi['Close'].iloc[0])*100
series_IXIC = (symbol_IXIC['Close']/symbol_IXIC['Close'].iloc[0])*100
series_SSEC = (symbol_SSEC['Close']/symbol_SSEC['Close'].iloc[0])*100
series_HSI = (symbol_HSI['Close']/symbol_HSI['Close'].iloc[0])*100
series_N225 = (symbol_N225['Close']/symbol_N225['Close'].iloc[0])*100
series_FTSE = (symbol_FTSE['Close']/symbol_FTSE['Close'].iloc[0])*100
series_FCHI = (symbol_FCHI['Close']/symbol_FCHI['Close'].iloc[0])*100
series_GDAXI = (symbol_GDAXI['Close']/symbol_GDAXI['Close'].iloc[0])*100
```

```python
fig, axes = plt.subplots(7,1,figsize=(10,50), dpi=100)


axes[0].plot(series_kospi, label='series_kospi')
axes[0].plot(series_IXIC, label='series_IXIC')
axes[0].set_title(' & ')
axes[0].set_xlabel('')
axes[0].set_ylabel('')
axes[0].legend()

axes[1].plot(series_kospi, label='series_kospi')
axes[1].plot(series_SSEC, label='series_SSEC')
axes[1].set_title(' & ')
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].legend()


axes[2].plot(series_kospi, label='series_kospi')
axes[2].plot(series_HSI, label='series_HSI')
axes[2].set_title(' & ')
axes[2].set_xlabel('')
axes[2].set_ylabel('')
axes[2].legend()


axes[3].plot(series_kospi, label='series_kospi')
axes[3].plot(series_N225, label='series_N225')
axes[3].set_title(' &  ')
axes[3].set_xlabel('')
axes[3].set_ylabel('')
axes[3].legend()


axes[4].plot(series_kospi, label='series_kospi')
axes[4].plot(series_FTSE, label='series_FTSE')
axes[4].set_title(' &  FTSE100')
axes[4].set_xlabel('')
axes[4].set_ylabel('')
axes[4].legend()


axes[5].plot(series_kospi, label='series_kospi')
axes[5].plot(series_FCHI, label='series_FCHI')
axes[5].set_title(' &  FCHI ')
axes[5].set_xlabel('')
axes[5].set_ylabel('')
axes[5].legend()


axes[6].plot(series_kospi, label='series_kospi')
axes[6].plot(series_GDAXI, label='series_GDAXI')
axes[6].set_title(' &  ')
axes[6].set_xlabel('')
axes[6].set_ylabel('')
axes[6].legend()

plt.tight_layout()
plt.show()
```
