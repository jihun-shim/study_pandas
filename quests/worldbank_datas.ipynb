```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'
import FinanceDataReader as fdr
mpl.rcParams['axes.unicode_minus'] = False
import seaborn as sns
import numpy as np
```

```python
def process_and_visualize(data, title, xlabel, ylabel):
    '''
    데이터를 처리하고 시각화하는 함수
    - data: 리스트 형태의 데이터
    - title: 그래프 제목
    - xlabel: x축 라벨
    - ylabel: y축 라벨
    '''

    # 1. 데이터를 DataFrame으로 변환
    df = pd.DataFrame([
        {
            'Country':entry['country']['value']
            , 'Year':int(entry['date'])
            , 'Value':entry['value']
        }
        for entry in data if entry ['value'] is not None
    ])
    
       # 2. 그래프 생성
    plt.figure(figsize=(10, 6))
    for country in df['Country'].unique():  # 국가별로 데이터를 나눔
        country_data = df[df['Country'] == country]  # 특정 국가의 데이터 선택
        plt.plot(country_data['Year'], country_data['Value'], marker='o', label=country)

    # 3. 그래프 꾸미기
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(title='Country', loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
