```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'
import FinanceDataReader as fdr
mpl.rcParams['axes.unicode_minus'] = False
import seaborn as sns
import numpy as np
```

```python
symbol_TSLA = fdr.DataReader(symbol = 'TSLA', start='2021-01-01', end='2024-12-01')
```

```python
symbol_TSLA
```

```python
symbol_TSLA['rolling5'] = symbol_TSLA['Close'].rolling(window=5).mean()
symbol_TSLA['rolling20'] = symbol_TSLA['Close'].rolling(window=20).mean()
symbol_TSLA['rolling60'] = symbol_TSLA['Close'].rolling(window=60).mean()
symbol_TSLA['rolling120'] = symbol_TSLA['Close'].rolling(window=120).mean()
```

```python
symbol_TSLA_rolling5 = symbol_TSLA['rolling5']
symbol_TSLA_rolling20 = symbol_TSLA['rolling20']
symbol_TSLA_rolling60 = symbol_TSLA['rolling60']
symbol_TSLA_rolling120 = symbol_TSLA['rolling120']
```

```python
print(f"이동평균선 5 최고치: {max(symbol_TSLA_rolling5)}, 최저치: {min(symbol_TSLA_rolling5)}")
print(f"이동평균선 20 최고치: {max(symbol_TSLA_rolling20)}, 최저치: {min(symbol_TSLA_rolling20)}")
print(f"이동평균선 60 최고치: {max(symbol_TSLA_rolling60)}, 최저치: {min(symbol_TSLA_rolling60)}")
print(f"이동평균선 120 최고치: {max(symbol_TSLA_rolling120)}, 최저치: {min(symbol_TSLA_rolling120)}")
```

```python
symbol_TSLA_rolling5
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,10), dpi=100)


axes[0,0].plot(symbol_TSLA_rolling5, label = '이동평균선 5')
axes[0,0].set_title('이동평균선 5')
axes[0,0].set_xlabel('날짜')
axes[0,0].set_ylabel('등락률')

axes[0,1].plot(symbol_TSLA_rolling20, label = '이동평균선 20')
axes[0,1].set_title('이동평균선 20')
axes[0,0].set_xlabel('날짜')
axes[0,0].set_ylabel('등락률')

axes[1,0].plot(symbol_TSLA_rolling60, label = '이동평균선 60')
axes[1,0].set_title('이동평균선 60')
axes[1,0].set_xlabel('날짜')
axes[1,0].set_ylabel('등락률')

axes[1,1].plot(symbol_TSLA_rolling120, label = '이동평균선 120')
axes[1,1].set_title('이동평균선 120')
axes[1,1].set_xlabel('날짜')
axes[1,1].set_ylabel('등락률')

plt.tight_layout()
plt.show()

print(f"이동평균선 5 최고치: {max(symbol_TSLA_rolling5)}, 최저치: {min(symbol_TSLA_rolling5)}")
print(f"이동평균선 20 최고치: {max(symbol_TSLA_rolling20)}, 최저치: {min(symbol_TSLA_rolling20)}")
print(f"이동평균선 60 최고치: {max(symbol_TSLA_rolling60)}, 최저치: {min(symbol_TSLA_rolling60)}")
print(f"이동평균선 120 최고치: {max(symbol_TSLA_rolling120)}, 최저치: {min(symbol_TSLA_rolling120)}")
```
