```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'
```

```python
import FinanceDataReader as fdr
symbol_kospi = fdr.DataReader('KS11','2023')
symbol_samsung_elec = fdr.DataReader('005930','2023')
```

```python
fig, axes = plt.subplots(4,1,figsize=(10,6), dpi = 100)

fig.suptitle('kospi and samsung_elec')

axes[0].plot(symbol_kospi['Close'])
axes[0].plot(symbol_samsung_elec['Close'])
axes[0].set_title('Close')
axes[0].set_xlabel('날짜')
axes[0].set_ylabel('지수')

axes[1].plot(symbol_kospi['High'])
axes[1].plot(symbol_samsung_elec['High'])
axes[1].set_title('High')
axes[1].set_xlabel('날짜')
axes[1].set_ylabel('지수')

axes[2].plot(symbol_kospi['Low'])
axes[2].plot(symbol_samsung_elec['Low'])
axes[2].set_title('Low')
axes[2].set_xlabel('날짜')
axes[2].set_ylabel('지수')

axes[3].plot(symbol_kospi['Volume'])
axes[3].plot(symbol_samsung_elec['Volume'])
axes[3].set_title('Volume')
axes[3].set_xlabel('날짜')
axes[3].set_ylabel('지수')


plt.legend()
plt.tight_layout()

plt.show()
```
