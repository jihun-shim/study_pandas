```python
import pandas as pd
```

```python
file_path = '../datasets/주식거래 수수료_20241217.xls'
compare_stock = pd.read_excel(file_path, header=2)
compare_stock
```

```python
compare_stock[['회사명','HTS.1','스마트폰.1']]
```

```python
compare_stock_sum = compare_stock[['HTS.1', '스마트폰.1']].sum(axis=1)
compare_stock_sum
```

```python
compare_stock['HTS1+스마트폰1']=compare_stock_sum
compare_stock
```

```python
new_compare_stock = compare_stock.sort_values(by=['HTS1+스마트폰1'], ascending=[True])
new_compare_stock
```

```python
new_compare_stock_list = new_compare_stock[['회사명','HTS1+스마트폰1']]
new_compare_stock_list
```
