```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'
import FinanceDataReader as fdr
mpl.rcParams['axes.unicode_minus'] = False
import numpy as np
```

```python
symbol_NFLX = fdr.DataReader(symbol = 'NFLX', start='2024-11-01', end='2024-12-19')
symbol_QUBT = fdr.DataReader(symbol = 'QUBT', start='2024-11-01', end='2024-12-19')
symbol_TSLA = fdr.DataReader(symbol = 'TSLA', start='2024-11-01', end='2024-12-19')
```

```python
symbol_NFLX['전일종가'] = symbol_NFLX['Close'].shift(periods=1)
symbol_NFLX['5일전_종가'] = symbol_NFLX['Close'].shift(periods=5)
symbol_NFLX['10일전_종가'] = symbol_NFLX['Close'].shift(periods=10)

symbol_QUBT['전일종가'] = symbol_QUBT['Close'].shift(periods=1)
symbol_QUBT['5일전_종가'] = symbol_QUBT['Close'].shift(periods=5)
symbol_QUBT['10일전_종가'] = symbol_QUBT['Close'].shift(periods=10)

symbol_TSLA['전일종가'] = symbol_TSLA['Close'].shift(periods=1)
symbol_TSLA['5일전_종가'] = symbol_TSLA['Close'].shift(periods=5)
symbol_TSLA['10일전_종가'] = symbol_TSLA['Close'].shift(periods=10)
```

```python
symbol_NFLX_one = (symbol_NFLX['Close']-symbol_NFLX['전일종가']) * 100
symbol_NFLX_five = (symbol_NFLX['Close']-symbol_NFLX['5일전_종가']) * 100
symbol_NFLX_ten = (symbol_NFLX['Close']-symbol_NFLX['10일전_종가']) * 100

symbol_QUBT_one = (symbol_QUBT['Close']-symbol_QUBT['전일종가']) * 100
symbol_QUBT_five = (symbol_QUBT['Close']-symbol_QUBT['5일전_종가']) * 100
symbol_QUBT_ten = (symbol_QUBT['Close']-symbol_QUBT['10일전_종가']) * 100

symbol_TSLA_one = (symbol_TSLA['Close']-symbol_TSLA['전일종가']) * 100
symbol_TSLA_five = (symbol_TSLA['Close']-symbol_TSLA['5일전_종가']) * 100
symbol_TSLA_ten = (symbol_TSLA['Close']-symbol_TSLA['10일전_종가']) * 100
```

```python
fig, axes = plt.subplots(3,1,figsize=(10,20), dpi=100)

axes[0].plot(symbol_NFLX_one, label='넷플릭스 전일종가')
axes[0].plot(symbol_QUBT_one, label='퀀텀 컴퓨팅 전일종가')
axes[0].plot(symbol_TSLA_one, label='테슬라 전일종가')
axes[0].set_title('1일전 등락률 비교')
axes[0].set_xlabel('날짜')
axes[0].set_ylabel('등락률')
axes[0].legend()

axes[1].plot(symbol_NFLX_five, label='넷플릭스 5일전 종가')
axes[1].plot(symbol_QUBT_five, label='퀀텀 컴퓨팅 5일전 종가')
axes[1].plot(symbol_TSLA_five, label='테슬라 5일전 종가')
axes[1].set_title('5일전 등락률 비교')
axes[1].set_xlabel('날짜')
axes[1].set_ylabel('등락률')
axes[1].legend()

axes[2].plot(symbol_NFLX_ten, label='넷플릭스 10일전 종가')
axes[2].plot(symbol_QUBT_ten, label='퀀텀 컴퓨팅 10일전 종가')
axes[2].plot(symbol_TSLA_ten, label='테슬라 10일전 종가')
axes[2].set_title('10일전 등락률 비교')
axes[2].set_xlabel('날짜')
axes[2].set_ylabel('등락률')
axes[2].legend()

plt.tight_layout()
plt.show()
```
